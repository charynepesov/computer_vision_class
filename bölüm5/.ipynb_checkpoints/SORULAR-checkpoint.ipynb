```python
# Örnek 5.2: OpenCV ile Video Kaydetme

video = cv2.VideoCapture("../DATA/videoplayback.mp4")

width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = video.get(cv2.CAP_PROP_FPS)

if video.isOpened() == False:
    print("Doğru gitmeyen bir şeyler var.")

vw = cv2.VideoWriter('videoplayback_son2FramedeYazi.mp4', cv2.VideoWriter_fourcc(*'XVID'), fps, (width, height), isColor=False)

frameler = []

while video.isOpened:
    ret, frame = video.read() #ilk resmi aldik

    if ret == True: #resmi basarili bir sekilde aldiysak
        cv2.imshow("framler", frame)
        frameler.append(frame)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    else:
        break
video.release()
cv2.destroyAllWindows()
```

```python
i = 0
for frame in frameler[::-1]:
    if i < 2:
        cv2.putText(frame, "Merhaba Dunya", (100,100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0))
        filename = "frame" + str(i) + ".jpg"
        cv2.imwrite(filename, frame)
        i += 1
        

for frame in frameler:        
    cv2.imshow("Duz", frame)
    vw.write(frame)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

vw.release()
cv2.destroyAllWindows()
```

```python
#SORU-2: videoA nin timeline kismini videoB deki timeline kismina ekle
```

```python
videoA = cv2.VideoCapture("ornek-1.mp4")
videoB = cv2.VideoCapture("ornek-2.mp4")

width = int(videoA.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(videoA.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = videoA.get(cv2.CAP_PROP_FPS)

if videoA.isOpened() == False:
    print("Doğru gitmeyen bir şeyler var.")

vw = cv2.VideoWriter('timelineDegistirme.mp4', cv2.VideoWriter_fourcc(*'XVID'), fps, (width, height), isColor=False)

timeline= []

while videoA.isOpened:
    ret, frame = videoA.read() #ilk resmi aldik

    if ret == True: #resmi basarili bir sekilde aldiysak
        cv2.imshow("framler", frame[:35,:400,:])
        timeline.append(frame[:35,:400,:])
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    else:
        break
videoA.release()
cv2.destroyAllWindows()

```

```python
i = len(timeline)
while videoB.isOpened:
    ret, frame = videoB.read() #ilk resmi aldik
    if ret == True: #resmi basarili bir sekilde aldiysak
        frame[:35,:400,:] = timeline[i-1]
        cv2.imshow("framler", frame)
        vw.write(frame)
        i -= 1
        if i == 0:
            break
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    else:
        break
videoB.release()
cv2.destroyAllWindows()
```
