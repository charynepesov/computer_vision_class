```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Örnek 4.6 - Canny Edge Detection

img = cv2.imread('../DATA/21_bird.jpg')

result1 = cv2.Canny(img,80,80)

medianValue = np.median(img) 
lowTH = int(max(0 ,0.7 * medianValue))
highTH = int(min(255, 1.3 * medianValue))

result2 = cv2.Canny(img,lowTH,highTH,L2gradient=True)

f, eksen = plt.subplots(1,3,figsize=(17,7))
eksen[0].imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
eksen[1].imshow(result1,cmap="gray")
eksen[2].imshow(result2,cmap="gray")
```
