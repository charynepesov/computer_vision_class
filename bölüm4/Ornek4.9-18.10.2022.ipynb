```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Örnek 4.9: Shi-Tomasi Corner Detection

img = cv2.imread('../DATA/22_real_chess.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
imgGray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

#goodFeaturesToTrack(image, maxCorners, qualityLevel, minDistance
#[, corners[, mask[, blockSize[, useHarrisDetector[, k]]]]]) -> corners

corners = cv2.goodFeaturesToTrack(imgGray,0,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),3,(255,0,0),-1)

f, eksen = plt.subplots(1,1,figsize=(8,5))
eksen.imshow(img)
```
