```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
resim = cv2.imread('../DATA/10_pubg.jpg')
resim = cv2.cvtColor(resim,cv2.COLOR_BGR2RGB)
resim.shape
```

```python
#Ornek 3.12 - gamma d√ºzeltme

yeni_resim = resim.copy()
for yuk in range(resim.shape[0]):
    for gen in range(resim.shape[1]):
        for kanal in range(resim.shape[2]):
            yeni_resim[yuk,gen,kanal] = np.clip(1.3*resim[yuk,gen,kanal]+15, 0, 255)
```

```python
#plt.imshow(yeni_resim)
f, eksen = plt.subplots(1,2)
eksen[0].imshow(resim)
eksen[1].imshow(yeni_resim)
```

```python
pencere = plt.figure()
pencere.add_subplot(2,2,1)
plt.imshow(resim)
pencere.add_subplot(2,2,2)
plt.imshow(yeni_resim)
```
