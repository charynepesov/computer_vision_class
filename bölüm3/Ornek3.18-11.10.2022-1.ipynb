```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Örnek 3.18 - Histogram Eğrisi
img = cv2.imread('../DATA/17_rainbow.jpg')
imgGray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

imgGray_hist = cv2.calcHist([imgGray],channels=[0],mask=None,histSize=[256],ranges=[0,256])
Eq = cv2.equalizeHist(imgGray)
Eq_hist = cv2.calcHist([Eq],channels=[0],mask=None,histSize=[256],ranges=[0,256])

f, eksen = plt.subplots(1,3,figsize=(17,5))
eksen[0].imshow(img)
eksen[0].title.set_text("Resim")
eksen[1].plot(imgGray_hist)
eksen[1].title.set_text("Orijinal Histogram")
eksen[2].plot(Eq_hist)
eksen[2].title.set_text("Eşitlemeden Sonra Histogram")
```
