```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# resimleri ekliyoruz
y = cv2.imread('../DATA/3_opencv-logo.png')
x = cv2.imread('../DATA/4_python_logo.png')
```

```python
#resimleri rgb ye çeviriyoruz 
x = cv2.cvtColor(x,cv2.COLOR_BGR2RGB)
y = cv2.cvtColor(y,cv2.COLOR_BGR2RGB)
```

```python
#resimlerin boyutuna bakıyoruz
x.shape
```

```python
y.shape
```

```python
#2.resmi 1.resmin boyutuna getiriyoruz
y = cv2.resize(y,(x.shape[1],x.shape[0]))
```

```python
# Örnek 3.14 - resim ekleme/birleştirme/çıkarma

sonuc1 = x + y
sonuc2 = cv2.add(x,y)
sonuc3 = cv2.addWeighted(x,0.9,y,0.9,gamma=10)
sonuc4 = cv2.subtract(x,y)
```

```python
f, eksen = plt.subplots(1,4)
eksen[0].imshow(sonuc1, cmap="gray")
eksen[1].imshow(sonuc2, cmap="gray")
eksen[2].imshow(sonuc3, cmap="gray")
eksen[3].imshow(sonuc4, cmap="gray")
```
