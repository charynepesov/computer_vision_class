```python
import cv2
import time
import numpy as np
# videonun 1dk*24=1444 frame sini alip vakit kismini biraz geriden alinacak ve yapistirilacak
# 2 video ile yapilabilir
```

```python
video1 = cv2.VideoCapture("ornek-1.mp4")
zamanV = cv2.VideoCapture("ornek-2.mp4")

width = int(video1.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video1.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = video1.get(cv2.CAP_PROP_FPS)

if zamanV.isOpened() == False:
    print("Doğru gitmeyen bir şeyler var.")
    
vw = cv2.VideoWriter('video_sahte.mp4', cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height), isColor=True)
sayac = 0
frame3 = np.zeros((height, width))
while zamanV.isOpened:
    ret, frameZ = zamanV.read()
    if ret == True:
        if(sayac >= (60 * fps) and sayac <= (120 * fps)):
            print(sayac)
            ret, frame1 = video1.read()
            if ret == True: #resmi basarili bir sekilde aldiysak
                time.sleep(1 / fps) #video nun gercek fps si ile oynat
                frame1[:50, :400] = frameZ[:50, :400]
                cv2.imshow('Video Dosya Okuma1', frame1)
                vw.write(frame1)
                if cv2.waitKey(10) & 0xFF == ord('q'):
                    break
        sayac += 1
    else:
        break

vw.release()
video1.release()
zamanV.release()
cv2.destroyAllWindows()
```
