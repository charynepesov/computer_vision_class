```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = plt.imread("../DATA/5_gazete.png")
imgGray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2.GaussianBlur(imgGray, ksize=5)
plt.imshow(img)
```

```python
Gx = cv2.Sobel(imgGray, cv2.CV_32F, dx=1, dy=0, ksize=3)
```

```python
Gy = cv2.Sobel(imgGray, cv2.CV_32F, dx=0, dy=1, ksize=3)
```

```python
G = cv2.sqrt((Gx * Gx) + (Gy * Gy))
```

```python
plt.imshow(G, cmap="gray")
```
