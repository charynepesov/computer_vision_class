```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread("DATA/1_ayasofya.jpg")
imgGray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
imgGray.shape
```

```python
ROI = imgGray[:5,:6]
ROI.shape
```

```python
print(ROI)
```

```python
plt.imshow(ROI, cmap="gray")
```

```python
sonuc = cv2.adaptiveThreshold(ROI, 200, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 5, 1)
print(sonuc)
```

```python
plt.imshow(sonuc, cmap="gray")
```
